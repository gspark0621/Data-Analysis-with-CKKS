```python
from scipy.io import arff
import pandas as pd
import matplotlib.pyplot as plt

# Hepta 데이터셋 불러오기
hepta_data, hepta_meta = arff.loadarff('./dataset/hepta.arff')
df_hepta = pd.DataFrame(hepta_data)

# Tetra 데이터셋 불러오기
tetra_data, tetra_meta = arff.loadarff('./dataset/tetra.arff')
df_tetra = pd.DataFrame(tetra_data)

# Twodiamonds 데이터셋 불러오기
Twodiamonds_data, Twodiamonds_meta = arff.loadarff('./dataset/Twodiamonds.arff')
df_Twodiamonds = pd.DataFrame(Twodiamonds_data)

# Lsun 데이터셋 불러오기
Lsun_data, Lsun_meta = arff.loadarff('./dataset/Lsun.arff')
df_Lsun = pd.DataFrame(Lsun_data)

# temp 데이터셋 불러오기
temp_data, temp_meta = arff.loadarff('./dataset/temp.arff')
df_temp = pd.DataFrame(temp_data)

# Hepta 데이터 시각화 (2차원 투영)
plt.figure(figsize=(7,5))
plt.scatter(df_hepta.iloc[:, 0], df_hepta.iloc[:, 1], c='blue', s=20)
plt.title('Hepta Dataset (2D projection)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

# Tetra 데이터 시각화 (2차원 투영)
plt.figure(figsize=(7,5))
plt.scatter(df_tetra.iloc[:, 0], df_tetra.iloc[:, 1], c='green', s=20)
plt.title('Tetra Dataset (2D projection)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

# Twodiamonds 데이터 시각화 (2차원 투영)
plt.figure(figsize=(7,5))
plt.scatter(df_Twodiamonds.iloc[:, 0], df_Twodiamonds.iloc[:, 1], c='green', s=20)
plt.title('Twodiamonds Dataset (2D projection)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

# Lsun 데이터 시각화 (2차원 투영)
plt.figure(figsize=(7,5))
plt.scatter(df_Lsun.iloc[:, 0], df_Lsun.iloc[:, 1], c='green', s=20)
plt.title('Lsun Dataset (2D projection)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()

# temp 데이터 시각화 (2차원 투영)
plt.figure(figsize=(7,5))
plt.scatter(df_temp.iloc[:, 0], df_temp.iloc[:, 1], c='green', s=20)
plt.title('temp Dataset (2D projection)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
